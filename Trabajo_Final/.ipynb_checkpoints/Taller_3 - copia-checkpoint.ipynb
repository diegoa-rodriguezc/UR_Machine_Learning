**Descripción de la información o dataset:**

Información de la actividad de 30.000 estudiantes en redes sociales.

Se presentan ciertas características sociodemográficas de los estudiantes (gradyear , gender, age and friends) y el resto de las variables son la frecuencia o el número de veces en que se encuentra la palabra en la red social de cada estudiante.

**Objetivo:**

Lograr realizar clusters usando solamente las frecuencias de las palabras. Así mismo, desde la película The Breakfast Club existen 5 tipos de jóvenes (princesses, brains, athletes, basket case, criminals) y la idea es conseguir segmentar a la sociedad de estudiantes en redes sociales. 

**Factores a tener en cuenta:**

* Recuerde que también existen métodos para saber en cuantos grupos segmentar (Siluette, Codo) y lo importante es lograr distinguir o interpretar cada uno de los grupos que usted segmento.
* Es importante realizar PCA con los 2 primeros componentes para observar si realmente este modelo da una idea de cuantos segmentos realizar (Tener en cuenta si la varianza explicada en los primeros 2 componentes es significativamente alto como para considerar esta metodología y explicar)
* No importa el algoritmo implementado para segmentación social, pero si es importante explicar el modelo usado y por qué cree que sirve (hierarchical Clustering Analysis, Kmeans , etc)

# Kmeans

```python
%%time

rango_n_grupos = range(1,15,1) #Cantidad de clústers a generar desde 1 hasta 15
inercias = []

for k in rango_n_grupos:
    modelo = KMeans(n_clusters = k, random_state=123)
    modelo.fit(df)
    inercias.append(modelo.inertia_)
```

```python
%%time

fig, ax = plt.subplots(figsize=(16,6))

# Gráfico Inercia K-Means
plt.scatter(rango_n_grupos, inercias, c = 'red', s=150, marker='o', alpha=0.5)
plt.plot(rango_n_grupos, inercias, lw=3, linestyle='-', label='Varianza Explicada')

# Títulos del Gráfico
ax.set(title='Gráfico Método del Codo', 
       xlabel='Número de Grupos, K-Means', 
       ylabel='Inercia')
      
plt.show()
```

```python
%%time

# Deshabilitar advertencias:
import warnings
warnings.filterwarnings('ignore')

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score

import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np

puntuaciones = []
all_n_grupos = []

for n_grupos in rango_n_grupos:
    try:
        if n_grupos>1:
            # Create a subplot 
            fig, ax1 = plt.subplots(figsize=(16,6))
            #fig.set_size_inches(18, 7)

            KMean = KMeans(n_clusters=n_grupos, random_state=10)
            KMean.fit(df)    
            etiquetas = KMean.predict(df)

            # The silhouette_score gives the average value for all the samples.
            # This gives a perspective into the density and separation of the formed
            # clusters
            score = silhouette_score(df, etiquetas)
            print(
                "Número de Grupos: ",
                n_grupos,
                "Silhouette Score: ",
                score,
            )
            puntuaciones.append(score)
            all_n_grupos.append(n_grupos)

            # Compute the silhouette scores for each sample
            sample_silhouette_values = silhouette_samples(df, etiquetas)

            y_lower = 10

            for i in range(n_grupos):
                # Aggregate the silhouette scores for samples belonging to
                # cluster i, and sort them
                ith_cluster_silhouette_values = sample_silhouette_values[etiquetas == i]

                ith_cluster_silhouette_values.sort()

                size_cluster_i = ith_cluster_silhouette_values.shape[0]
                y_upper = y_lower + size_cluster_i

                color = cm.nipy_spectral(float(i) / n_grupos)
                ax1.fill_betweenx(
                    np.arange(y_lower, y_upper),
                    0,
                    ith_cluster_silhouette_values,
                    facecolor=color,
                    edgecolor=color,
                    alpha=0.7,
                )
                # Label the silhouette plots with their cluster numbers at the middle
                ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))
                # Compute the new y_lower for next plot
                y_lower = y_upper + 10  # 10 for the 0 samples

            ax1.set_title("Silhouette para varios clúster.")
            ax1.set_xlabel("Silhouette Score")
            ax1.set_ylabel("Grupos/Clúster")

            # The vertical line for average silhouette score of all the values
            ax1.axvline(x=score, color="red", linestyle="--")

            ax1.set_yticks([])  # Clear the yaxis labels / ticks
            ax1.set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])

            plt.suptitle(
                "Análisis Silhouette para Grupos/Clúster = %d"
                % n_grupos,
                fontsize=14,
                fontweight="bold",
            )
    except:
        pass

plt.show()
```

```python
maximo = np.max(puntuaciones)
print("El Silhouette Score maximo es: ", maximo)
print("El numero de Grupos sugerido por el modelo es:",all_n_grupos[puntuaciones.index(maximo)])
```

```python
# Se crea un df usando solamente las frecuencias de las palabras
#features = df.iloc[:,3:]
#features.head(10)
interests = data.iloc[:,4:39]
interests
```

```python
from sklearn import preprocessing
# Preprocesado utilizando Sklearn, StandardScaler:
# Estandarización de la tabla:
scaler = preprocessing.StandardScaler().fit(interests)

base_escalados = scaler.transform(interests)
base_escalados
```

```python
rango_n_grupos = range(1,5,1)
inercias = []

fig, ax = plt.subplots(figsize=(5,5))

for k in rango_n_grupos:
    try:
        modelo = KMeans(n_clusters = k, random_state=10)
        modelo.fit(base_escalados.T)
        inercias.append(modelo.inertia_)
    except:
        pass
    
# Gráfico Inercia K-Means
plt.scatter(rango_n_grupos, inercias, c = 'red', s=150, marker='o', alpha=0.5)
plt.plot(rango_n_grupos, inercias, lw=3, linestyle='-', label='Varianza Explicada')

# Títulos del Gráfico
ax.set(title='Gráfico Método del Codo', 
       xlabel='Número de Grupos, K-Means', 
       ylabel='Inercia')
      
plt.show()
```
